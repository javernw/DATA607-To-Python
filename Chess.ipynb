## Tidy Data

Data imported is cleaned and converted to a dataframe which will make data manipulation possible was we move forward.

```python
 # use of regex to remove dashes
chess_file = re.sub("[-]{2,}", " ", chess_file)

# using "|" as seperator to create columns for dataframe. Each row ends via the newline character.
chess_file = pd.DataFrame([x.split('|') for x in chess_file.split('\n')])

#remove rows and columns that are empty
chess_file = chess_file[chess_file.index % 3 != 0] 
chess_file = chess_file.reset_index(drop=True)

#remove first two columns as they consist of the column names
chess_file.drop(chess_file.index[[0, 1]], inplace=True)
chess_file.drop(chess_file.columns[10], axis=1, inplace=True)
chess_file = chess_file.reset_index(drop=True)

chess_file.columns = ["PairNum", "Player Name USCF ID / Rtg (Pre->Post)", "TotalNum", "Round1", "Round2", "Round3", "Round4", "Round5", "Round6", "Round7"]
chess_file
```
